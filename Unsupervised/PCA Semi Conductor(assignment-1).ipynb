```python
import pandas as pd
import numpy as np
```

```python
urldata="http://archive.ics.uci.edu/ml/machine-learning-databases/secom/secom.data"
data = pd.read_csv(urldata,sep=" ",header=None)
data.head()
```

```python
a=list(data.isna().sum())
a.sort()
print(a[len(a)-6:len(a)])
```

```python
for i in list(data):
    data[i] = data[i].fillna(data[i].mean())
```

```python
data.head(10)
```

```python
def mean(x):
    return x-np.mean(x)
data=data.apply(mean)
```

```python
data.head()
```

# Standardizing data

```python
for i in list(data):
    mean = data[i].mean()
    std = data[i].std()
    if std>0:
        data[i] = data[i].apply(lambda x:(x-mean)/std)
```

```python
data.head()
```

```python
data.describe()
```

```python
data.head()
```

# 2.Compute the covariance matrix(Σ):

```python
x=data.values
y=(np.dot(np.transpose(x),x)/len(x))
print(y.shape)
sigma = (np.matmul(np.transpose(x),x))/len(x)
sigma.shape
```

# 3.Calculate the Eigen Values and Eigen Vectors

```python
from scipy.linalg import eigh
values,vectors=eigh(sigma)
print(vectors.shape)
print(values.shape)
```

```python
print(values[589])
```

# 4.Plot the amount of variance that is preserved across each principal component and try to figure out the optimal number of principal components that we can get.

```python
var=values/np.sum(values)*100
components= list(range(len(var)))
cumulative_var = np.cumsum(var)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(cumulative_var)
plt.xlabel("Components")
plt.ylabel("cumulative variance")
plt.title("Components VS cumulative variance")
plt.show()
```
